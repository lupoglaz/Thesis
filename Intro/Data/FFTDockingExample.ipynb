```python
def loadLigand(num):
    ligandIm = misc.imread('ligandS%d.png'%num)
    ligand = np.zeros((ligandIm.shape[0],ligandIm.shape[1]))
    for i in range(0,ligandIm.shape[0]):
        for j in range(0,ligandIm.shape[1]):
            if ligandIm[i,j,0]==0 and ligandIm[i,j,1]==0 and ligandIm[i,j,2]==0:
                ligand[i,j]=1
            if ligandIm[i,j,0]==255 and ligandIm[i,j,1]==0 and ligandIm[i,j,2]==0:
                ligand[i,j]=0.5
    return ligand

def loadReceptor():
    #receptorIm = misc.imread('receptor.png')
    receptorIm = misc.imread('receptorSimpleNew.png')
    receptor = np.zeros((receptorIm.shape[0],receptorIm.shape[1]))
    for i in range(0,receptorIm.shape[0]):
        for j in range(0,receptorIm.shape[1]):
            if receptorIm[i,j,0]==0 and receptorIm[i,j,1]==0 and receptorIm[i,j,2]==0:
                receptor[i,j]=1
            if receptorIm[i,j,0]==255 and receptorIm[i,j,1]==0 and receptorIm[i,j,2]==0:
                receptor[i,j]=-10
    return receptor
```

```python
def dockLigand(Receptor, Ligand):
    fftLigand = fftp.fft2(Ligand)
    fftReceptor = fftp.fft2(Receptor)
    fftC = np.conjugate(fftReceptor)*fftLigand
    C = fftp.ifft2(fftC)
    return unravel_index(C.argmax(), C.shape), np.abs(C.max()), fftLigand, fftC, C

def shiftMolecule(coords, Ligand, Receptor):
    sLigand = Ligand.copy()
    img = np.zeros((Receptor.shape[0], Receptor.shape[1]))
    for i in range(Ligand.shape[0]):
        for j in range(Ligand.shape[1]):
            x = (i-coords[0])
            y = (j-coords[1])
            if x<0: x+=Ligand.shape[0]
            if y<0: y+=Ligand.shape[1]
            sLigand[x,y]=Ligand[i,j]
            sLigand[x,y]+=Receptor[x,y]
            if sLigand[x,y]!=0:
                img[x,y]=0.5
                #img[x,y,0]=255
                #img[x,y,1]=0
                #img[x,y,2]=0
            if Receptor[x,y]!=0:
                img[x,y]=1.0
                #img[x,y,0]=0
                #img[x,y,1]=255
                #img[x,y,2]=0
            #img[x,y,3]=1
    return img

def prepareLigandToPlot(Ligand):
    img = np.zeros((Ligand.shape[0], Ligand.shape[1]))
    for i in range(Ligand.shape[0]):
        for j in range(Ligand.shape[1]):
            if Ligand[i,j]==1:
                img[i,j]=1.0
                #img[i,j,0]=255
                #img[i,j,1]=0
                #img[i,j,2]=0
            if Ligand[i,j]==0.5:
                img[i,j]=0.5
                #img[i,j,0]=150
                #img[i,j,1]=20
                #img[i,j,2]=20
            #img[i,j,3]=0
    return img

def prepareReceptorToPlot(Receptor):
    img = np.zeros((Receptor.shape[0], Receptor.shape[1]))
    for i in range(Receptor.shape[0]):
        for j in range(Receptor.shape[1]):
            if Receptor[i,j]==1:
                img[i,j]=1.0
                #img[i,j,0]=255
                #img[i,j,1]=0
                #img[i,j,2]=0
            if Receptor[i,j]<0:
                img[i,j]=0.5
                #img[i,j,0]=150
                #img[i,j,1]=20
                #img[i,j,2]=20
            #img[i,j,3]=0
    return img

def prepateComplexArray(C):
    def g(x):
        return (1.0-1/(1+x**2))**0.2
    def Hcomplex(z):# computes the hue corresponding to the complex number z
        H=np.angle(z)/(2*np.pi)+1
        return np.mod(H,1)
    CA=fftshift(C)
    img = np.zeros((CA.shape[0], CA.shape[1],3))
    for i in range(CA.shape[0]):
        for j in range(CA.shape[1]):
            z=CA[i,j]
            H=Hcomplex(z)
            V=g(np.absolute(z))
            S=0.9
            img[i,j,0]=H
            img[i,j,1]=S
            img[i,j,2]=10.0
    #print img[100,100,:]
    return hsv_to_rgb(img)
```
